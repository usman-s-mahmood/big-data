```python
query1="select * from ucp.dept"
```

```python
# executing cursor 
cursor.execute(query1) 
  
# display all records 
table = cursor.fetchall() 
  
# fetch all columns 
print('\n Table Data:') 
for row in table: 
    print(row[0], end=" ") 
    print(row[1], end="\n") 
    
```
