```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Using Bar Plot

```python
x = ['python', 'c', 'c++', 'java', 'javascript']
y = [95, 70, 75, 80, 85]
z = [50, 20, 30, 45, 45]
p = np.arange(len(x)) # indexes for x list
width = 0.2
p1 = [j+width for j in p]
plt.xlabel('Languages', fontsize=14)
plt.ylabel('popularity percent', fontsize=14)
plt.title('Top 5 programming languages')
plt.bar(
    p, y, 
    width=0.2, 
    color='orange',
    align = 'center',
    edgecolor = 'black',
    linewidth = 5,
    linestyle = '', # : for dotted linestyle,
    alpha=0.9, # makes the graph lite and blur
    label='Big Data Analytics' # use plt.legend to show this label
)
# plt.legend()
# plt.show()
plt.bar(
    p1, z, 
    width=0.2, 
    color='green',
    align = 'center',
    edgecolor = 'black',
    linewidth = 5,
    linestyle = '', # : for dotted linestyle,
    alpha=0.9, # makes the graph lite and blur
    label='Big Data Analytics 2' # use plt.legend to show this label
)
plt.xticks(p+width/2, x, rotation=5)
plt.legend(loc='center')
plt.savefig('barplot.png', dpi=2000)
plt.show()
```

# Scatter Plot

```python
x = [11, 31, 15, 72, 89]
y = [20, 14, 6, 8, 12]
colors = ['r', 'y', 'g', 'b', 'r']
colors = [80, 29, 51, 70, 40]
plt.scatter(
    x, y, 
    c=colors, 
    cmap = 'Accent_r',
    s=[900, 800, 600, 600, 1200], 
    alpha=0.7,
    marker='*',
    edgecolor = 'black'
)
t = plt.colorbar()
t.set_label('ColorBar')
plt.title('Scatter plot', fontsize = 25)
plt.xlabel('Day')
plt.ylabel('Number')
plt.savefig('scatterplot.png', dpi=2000)
plt.show()
```

# Matplotlib Histogram Plot

```python
x = np.random.randint(10, 60, (30))
y = np.random.randint(30, 90, (30))
```

```python
plt.hist(
    x, color='g',
    edgecolor='black', 
    cumulative=1, # for reversing graph axis
    align = 'mid',
    orientation='horizontal',
    rwidth = 0.8,
    label = 'Understanding Histograms'
    # log = True # for log values in table
)
plt.axvline(45, color='#ad0000')
plt.title('Matplotlib histogram')
plt.legend()
plt.grid()
plt.savefig('histogramplot.png', dpi=2000)
plt.show()
```

# Matplotlib Pie Plot

```python
x = [43, 43, 39, 31, 27]
y = ['Web development' , 'Data Science', 'Machine Learning', 'Script Automation', 'Software testing']
ex = [0.3, 0.3, 0.4, 0.3, 0.5]
plt.pie(
    x, labels=y, #explode=[]
    colors = ['g', 'r', '#ad0000', 'silver', 'blue'],
    autopct = '%0.2f%%', # for showing percentages
    shadow = True,
    radius = 1,
    labeldistance = 1.3,
    startangle = 360,
    textprops = {
        'fontsize': 15,
        'color': 'black'
    },
    wedgeprops = {
        'linewidth': 5.5,
        'edgecolor': '#c71e55'
    }
)
plt.title('Use of python in different fields')
plt.legend(loc=9)
plt.savefig('pieplot.png', dpi=2000)
plt.show()
```

# Matplotlib Stem plot

```python
x2 = [21, 31, 25, 37, 28]
y2 = ['Embeded Development', 'Desktop Development', 'Network Programming', 'Software Prototyping', 'Educational Purposes']
plt.stem(
    y2, x2,
    linefmt = ':',
    markerfmt='r*',
    bottom=36,
    basefmt = 'g',
    label = 'python',
    orientation = 'horizontal'
)
plt.title('Use of python in different fields')
plt.legend()
plt.savefig('stemplot.png', dpi=2000)
plt.show()
```

# matplotlib box plot

```python
x = [10, 20, 30, 40, 50, 60, 70, 100]
x1 = [5, 25, 20, 20, 40, 65, 60, 90]
y = [x, x1]
plt.boxplot(y, labels=['python', 'c++'])
plt.savefig('histogram.png', dpi=2000)
plt.show()
```

# matplotlib stack plot

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 3, 5, 10]
area1 = [11, 12, 11, 10, 13]
area2 = [20, 13, 18, 19, 21]
lbls = ['area1', 'area2', 'area3']
plt.stackplot(
    x, y, 
    area1, area2, 
    labels=lbls, 
    colors = [
        'maroon', 'green', 'silver', 'green'
    ]
)
plt.title('python')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()
plt.legend()
plt.savefig('stackplot.png', dpi=2000) # use facecolor to change the boundary color of picture, format can be .png, .pdf, .jpeg, .jpg, .gif
plt.show()
```
