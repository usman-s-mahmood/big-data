```python
mtl.plot([1,2,3],[5,10,15])
mtl.xlabel("x value")
mtl.ylabel("y value")
mtl.title("plot")
```

```python
x=[1,2,3]
y=[5,10,15]
mtl.plot(x,y,"go")
mtl.xlabel("x value")
mtl.ylabel("y value")
mtl.title("plot")
```

```python
x=[1,2,3]
y=[5,10,15]
mtl.subplot(x,y,"go")
mtl.xlabel("x value")
mtl.ylabel("y value")
mtl.title("subplot")
x=[1,2,3]
y=[5,10,15]
mtl.subplot(x,y,"r^")
mtl.xlabel("x value")
mtl.ylabel("y value")
mtl.title("subplot2")
```
